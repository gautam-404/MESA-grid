```python
import glob
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def get_velocity(history_file):
    history = pd.read_table(history_file, skiprows=5, sep='\s+')
    t = list(history['star_age'])
    v_surf = list(history['surf_avg_v_rot'])
    return t, v_surf
```

```python
histories = sorted(glob.glob("grid_archive_old/histories/*"))
t, vsurf = [], []
plt.figure(figsize=(10,7))
for history in histories:
    t_, v_surf_ = get_velocity(history)
    t.append(t_)
    vsurf.append(v_surf_)
    plt.plot(t_, v_surf_, label=f"{round(v_surf_[0])} km/s")
plt.xlabel("Age [yr]")
plt.ylabel("Surface velocity [km/s]")
# plt.xlim(0, t[0][-60])
plt.legend(title="M = 1.7 MSun, initial_Z = 0.017\n\nInitial velocity", loc='lower right')
plt.show()
```
